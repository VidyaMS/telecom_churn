```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
churn_data = pd.read_csv("data_telecom_churn.csv")
```

```python
def file_details(filename):

    print(filename.head())
    print("\n")
    print(filename.info())
    print("\n")
    print("Null values:")
    print(filename.isnull().sum())
```

```python
file_details(churn_data)
```

```python
plt.figure(figsize = (15,5))
f, (ax1,ax2) = plt.subplots(1, 2, figsize = (15,5))

churn_data[churn_data['Churn'] == 1]['International plan'].value_counts().plot(kind = 'bar', ax = ax1)
ax1.set_title("Churn = True")
ax1.set_xlabel("International Plan")
churn_data[churn_data['Churn'] == 0]['International plan'].value_counts().plot(kind = 'bar', ax = ax2)
ax2.set_title("Churn = False")
ax2.set_xlabel("International Plan")
```

```python
plt.figure(figsize = (15,5))
f, (ax1,ax2) = plt.subplots(1, 2, figsize = (15,5))

churn_data[churn_data['Churn'] == 1]['Voice mail plan'].value_counts().plot(kind = 'bar', ax = ax1)
ax1.set_title("Churn = True")
ax1.set_xlabel("Voice mail Plan")
churn_data[churn_data['Churn'] == 0]['Voice mail plan'].value_counts().plot(kind = 'bar', ax = ax2)
ax2.set_title("Churn = False")
ax2.set_xlabel("Voice mail Plan")
```

```python
plt.figure(figsize = (20,5))
f, (ax1,ax2) = plt.subplots(1, 2, figsize = (15,5))

churn_data[churn_data['Churn'] == 1]['Area code'].value_counts().plot(kind = 'bar', ax = ax1)
ax1.set_title("Churn = True")
ax1.set_xlabel("Area code")
churn_data[churn_data['Churn'] == 0]['Area code'].value_counts().plot(kind = 'bar', ax = ax2)
ax2.set_title("Churn = False")
ax2.set_xlabel("Area code")
```

Summary : 
1. No of records - 3333  
2. Non numeric columns - (State , International plan , Voice mail plan).    
3. No missing values.
4. @ 500 records with churn == true . Imbalanced data set - 14% 
5. Columns that are continuous variables - 15 
6. Among the continous variables, there doesn't seem to be linearity among them and each of them appear to be Guassian.
7. For those who 'churned', the mean day minutes and charges are higher than 'non-churned'. Not much difference observed in night minutes and international minutes and charges.Higher number of outliers in non-churned.  
8. 40% opted for International Plan among those who churned, while only 6% opted among the non-churned.
9. 20% opted for Voice mail plan among those who churned, where as among the 'non-churned' , 40% opted for voice mail plan. 
10. More number of voice mail messages for those 'non churned'.
